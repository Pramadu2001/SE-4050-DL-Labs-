```python
import matplotlib.pyplot as plt
```

```python
exp_data = np.random.exponential(scale=1.0, size=(100000, 1))
uniform_data = np.random.uniform(low=0, high=5, size=(100000, 1))
normal_data = np.random.normal(loc=2.5, scale=1.0, size=(100000, 1))

# 2. Plot histograms
plt.figure(figsize=(10, 6))

plt.hist(exp_data, bins=100, alpha=0.5, label='Exponential', density=True)
plt.hist(uniform_data, bins=100, alpha=0.5, label='Uniform', density=True)
plt.hist(normal_data, bins=100, alpha=0.5, label='Normal', density=True)

plt.title('Exponential vs Uniform vs Normal Distributions')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.axis([-2.5,10,0,1.1])
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D  # required for 3D plotting

# 1. Create meshgrid for X and Y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# 2. Compute Z
Z = X**2 + Y**2

# 3. Plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Surface plot
ax.plot_surface(X, Y, Z, cmap='viridis')

# Labels
ax.set_title(r'$Z = X^2 + Y^2$')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a dataset from seaborn's example data
df = sns.load_dataset('titanic')

# Select numeric columns for correlation analysis
# Exclude 'survived' and 'pclass' as they are categorical representations
stats = ["age", "sibsp", "parch", "fare"]
data = df[stats].dropna()

# Calculate Pearson and Spearman correlations
pearson_corr = data.corr(method='pearson')
spearman_corr = data.corr(method='spearman')

# Plot Pearson heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Pearson Correlation (Titanic Data)")
plt.show()

# Plot Spearman heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Spearman Correlation (Titanic Data)")
plt.show()
```
